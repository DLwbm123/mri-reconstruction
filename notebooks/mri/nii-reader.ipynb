```python
def read(fname):
    img = nibabel.load(fname)
    return np.array(img.get_fdata())
```

```python
im = read('/home/act65/Downloads/sub-01_inplaneT2.nii')
```

```python
plt.figure(figsize=(16,16))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.axis('off')
    plt.imshow(im[:, :, i], interpolation='nearest', cmap='gray')
```
