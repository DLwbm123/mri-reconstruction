```python
def read(fname):
    img = nibabel.load(fname)
    return np.array(img.get_fdata())
```

```python
im = read('/home/act65/Downloads/ds108_R2.0.0/sub-02/anat/sub-02_T1w.nii.gz')
im.shape
```

```python
plt.figure(figsize=(16,16))
for i in range(50):
    plt.subplot(10,5,i+1)
    plt.axis('off')
    plt.imshow(im[:, :, i], interpolation='nearest', cmap='gray')
```
